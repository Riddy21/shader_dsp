```python
import wave
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Open the wave file
wave_file = wave.open('test.wav', 'rb')

# Step 2: Extract parameters
n_channels = wave_file.getnchannels()
sampwidth = wave_file.getsampwidth()
framerate = wave_file.getframerate()
n_frames = wave_file.getnframes()

# Step 3: Read frames and convert to numpy array
frames = wave_file.readframes(n_frames)
wave_file.close()

# Convert frames to numpy array
audio_data = np.frombuffer(frames, dtype=np.int16)

# Separate channels
channel_1 = audio_data[0::2] / (2*32768) + 0.5
channel_2 = audio_data[1::2]

# Step 4: Generate time axis
time = np.linspace(0, len(channel_1) / framerate, num=len(channel_1))

# Step 5: Plot the waveform for each channel
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(time, channel_1)
plt.title('Waveform of media/test.wav - Channel 1')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(time, channel_2)
plt.title('Waveform of media/test.wav - Channel 2')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
